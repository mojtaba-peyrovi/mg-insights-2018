```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
sales2016=pd.read_excel("sales2016.xlsx")
```

```python
sales2016
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x='nationality',data=sales2016,hue='status')
sns.set_context('poster')
```

```python
fp = sales2016.pivot_table(index='month',columns='year',values='revenueTHB',aggfunc=np.sum)
```

```python
fp
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(fp,cmap='hot',linecolor='white',linewidth=1,annot=True,fmt='g')
```

```python
fp2 = sales2016.pivot_table(index='Insurer',columns='year',values='revenueTHB',aggfunc=np.sum)
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(fp2,cmap='YlGn',linecolor='white',linewidth=1,annot=True,fmt='g')
```
